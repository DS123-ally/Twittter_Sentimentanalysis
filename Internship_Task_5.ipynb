```python
import pandas as pd
```

```python
df=pd.read_csv("/content/twitter_training.csv", header=None, names=['ID', 'Game', 'Sentiment', 'Tweet'])
```

```python
df
```

```python
# Remove rows with missing tweet text
df_cleaned = df.dropna(subset=["Tweet"])

# Remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates
print("Cleaned dataset shape:", df_cleaned.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from  wordcloud import WordCloud
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Sentiment",data=df_cleaned,color="red")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
top_topics=df_cleaned["Game"].value_counts().head(10)
print(top_topics)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y="Game", data=df_cleaned, color="yellow", order=top_topics.index)
plt.title("Top 10 Games")
plt.ylabel("Game")
plt.xlabel("Tweet Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
